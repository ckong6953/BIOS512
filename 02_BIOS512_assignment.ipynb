Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.res=600, repr.plot.width=7,repr.plotheight=3)

library('scales')

p = ggplot(penguins, aes(x=body_mass_g))

p = p + geom_histogram(bins = 37, color="#1f78b4", fill="#1f78b4", alpha = 0.6, na.rm = TRUE)

p = p + facet_wrap(~year,nrow=3)

p = p + labs(title="Histograms of penguin body mass for measurements taken 2007-2009", x = "Body mass in grams", y="Number of penguins")

p = p + theme_minimal()

p = p + theme(strip.text = element_text(hjust = 0.95), plot.margin = margin (0, 25, 0, 0))

p = p + scale_x_continuous(labels = comma)

p
```

Is there a difference between the body mass of male and female penguis? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to see compare within each species-group.**

```r
options(repr.plot.width = 6, repr.plot.height = 5)

p = ggplot(penguins, aes(x=body_mass_g, y=sex))

p = p + geom_point(position = position_jitter(height=0.2), alpha=0.7, color="grey40", na.rm = TRUE)

p = p + facet_wrap(~species, nrow = 3)

p = p + labs(x="Body mass in grams")

p = p + theme_minimal()

p = p + theme(strip.text = element_text(hjust = 0.05))

p = p + scale_x_continuous(breaks = c(3500, 4500, 5500), labels=comma)

p 
```
